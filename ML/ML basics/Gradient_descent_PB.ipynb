```python
def gradient_descent(x,y,iterations,learning_rate): #x and y are arrays
    m_curr = b_curr =  0
    iterations = iterations
    n = len(x)
    lr = learning_rate
    for i in range(iterations):
        y_pred = m_curr*x + b_curr
        cost_f = (1/n)*(sum([val**2 for val in (y_pred-y)]))
        dm = (-2/n)*(y - y_pred)*x
        db = (-2/n)*(y - y_pred)
        print("{} is the cost for {} iteration,value of y_pred is {} and y actual is {}".format(cost_f,i,y_pred,y))
        m_curr = m_curr - lr*dm
        b_curr = b_curr - lr*dm
        
        
        
        
    
```
